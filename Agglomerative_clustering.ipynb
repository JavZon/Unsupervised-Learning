```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=70,centers=3,random_state=21)
```

```python
X[:3]
```

```python
# y - assigns each point to a cluster
y
```

```python
from sklearn.cluster import AgglomerativeClustering

agg=AgglomerativeClustering(n_clusters=3)
agg.fit_predict(X)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X[:,0],X[:,1],c=agg.labels_,cmap='rainbow', marker='o')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the confusion matrix
y_pred = agg.fit_predict(X)
conf_matrix = confusion_matrix(y, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Agglomerative Clustering Labels")
plt.ylabel("True Labels (make_blobs)")
plt.show()
```
