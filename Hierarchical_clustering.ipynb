```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='/content/Mall_Customers.csv'
df=pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df2=df.sample(30)
```

```python
df2
```

We only take 2 columns : **Annual income** and **Spending score** -  to classify the customers' type

```python
df3=df2.drop(columns=['CustomerID','Gender','Age'])
```

```python
df3.head()
```

```python
df3=df3.sort_index(ignore_index=True)
df3.head()
```

```python
X=df3.values
X[:5]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hier_cluster=AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='average')
hier_cluster.fit(X_scaled)
```

```python
hier_cluster.labels_
```

```python
df3['cluster']=hier_cluster.labels_
```

```python
df3.head()
```

Now we see the mean of each cluster

```python
df3.groupby('cluster').mean()
```

From the results, we can draw the following conclusions:

**Cluster 0**

Customers in this cluster have a significantly higher annual income compared to the other clusters.They have a relatively low spending score, suggesting that they may be more conservative in their spending habits or prioritize saving over spending.

**Cluster 1**

Customers in this cluster have a moderate annual income.
They have a moderate spending score, indicating a balanced approach to spending.

**Cluster 2**

Customers in this cluster have a high annual income.
They also have a high spending score, suggesting that they are likely to be affluent and spend freely.
Based on these observations, we can segment the customers into three distinct groups with different spending behaviors and income levels.
